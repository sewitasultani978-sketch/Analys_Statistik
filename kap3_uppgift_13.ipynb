```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
# Läsa in datasetet 
df = pd.read_excel("hr_employee_data.xlsx")
 
```

```python
df.head()
```

```python
df.describe()
```

```python
# Här räknar jag ut medelvärdet av satisfaction_level och det högsta och lägsta värdet av last_evaluation. 
df["satisfaction_level"].mean()
df["last_evaluation"].max()
df["last_evaluation"].min()
```

```python
# Här använder jag groupby för att räkna ut medelvärdet av satisfaction_level i varje avdelning. 
df.groupby("Department")["satisfaction_level"].mean()
```

```python
# Här räknar jag ut korrelationen mellan satisfaction_level och last_evaluation. 
df["satisfaction_level"].corr(df["last_evaluation"])
```

```python
# Detta diagram visar hur många anställda som har slutat i varje avdelning. 
left_counts = df.groupby("Department")["left"].sum()
plt.figure(figsize=(10,6))
sns.barplot(x=left_counts.index, y=left_counts.values)
plt.title("Antal anställda som har slutat i varje avdelning")
plt.xlabel("Department")
plt.ylabel("Antal som har slutat")
plt.xticks(rotation=45)
plt.show()
```
